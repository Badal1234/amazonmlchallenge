```python
import pandas as pd
import re
import string
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
df = pd.read_csv('train1.csv',header = None)
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
name = []
a = df.loc[0].dropna()
a[0]

for i in range(len(df)):
  s = df.loc[i].dropna()[0]
  res = re.sub('[%s]' % re.escape(string.punctuation), '' , s)
  res = re.sub(r'\w*\d\w*','',res.lower())
  res = lemmatize_words(res)
  name.append(res)

print(name)
```

```python
a = df.loc[1].dropna()
description = []
def findDescription(arr):
  stri = []
  for i in range(1,len(arr)):
    if(str(a[i])[0] == '['):
      break
    else:
      stri.append(a[i])
  return ''.join(dict.fromkeys(stri))
for i in range(len(df[:20000])):
  s = df.loc[i].dropna()
  s = findDescription(s)
  res = re.sub('[%s]' % re.escape(string.punctuation), '' , s)
  res = re.sub(r'\w*\d\w*','',res.lower())
  res = lemmatize_words(res)
  description.append(res)
print(description)



```

```python
bulletpoints = []

def extractBulletPoints(a):
  print(a)
  start = 0
  end = 0
  for i in range(len(a)):
    if(a[i][0] == '['):
      start = i
      break
  for i in range(len(a)):
    if((a[i])[-1] == ']'):
      end = i
      break
  if(start != end):
    return a[start: end+1]
  return []
for i in range(len(df)):
  s = df.loc[i].dropna()
  res = extractBulletPoints(s)
  if(len(res)):
    res = ''.join(dict.fromkeys(res))
    res = re.sub('[%s]' % re.escape(string.punctuation), '' , res)
    res = re.sub(r'\w*\d\w*','',res.lower())
    res = lemmatize_words(res)
    bulletpoints.append(res)
  else:
    removeIndexes.append(i)
print(bulletpoints)
  


    
```
